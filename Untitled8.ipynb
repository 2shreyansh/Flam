```python

!pip install numpy pandas scipy matplotlib --quiet
import numpy as np
import pandas as pd
from math import sin, cos, exp, pi
from scipy.optimize import differential_evolution, minimize
import matplotlib.pyplot as plt

CSV_FILE = "/xy_data.csv"
T_MIN, T_MAX = 6.0, 60.0
BOUNDS = [(0.0, 50.0), (-0.05, 0.05), (0.0, 100.0)]  # Î¸ (deg), M, X
def load_data(path):
    df = pd.read_csv(path)
    if not {'x','y'}.issubset(df.columns):
        df = df.iloc[:, :2]
        df.columns = ['x','y']
    return df[['x','y']].to_numpy()

def predict(theta_rad, M, X, tvals):
    t = tvals
    term = np.exp(M * np.abs(t)) * np.sin(0.3 * t)
    x = t * np.cos(theta_rad) - term * np.sin(theta_rad) + X
    y = 42 + t * np.sin(theta_rad) + term * np.cos(theta_rad)
    return np.vstack([x, y]).T

def align_and_l1(params, obs, tvals):
    th_deg, M, X = params
    th = np.deg2rad(th_deg)
    pred = predict(th, M, X, tvals)
    c_obs, c_pred = obs.mean(axis=0), pred.mean(axis=0)
    ang_obs = np.arctan2(obs[:,1]-c_obs[1], obs[:,0]-c_obs[0])
    ang_pred = np.arctan2(pred[:,1]-c_pred[1], pred[:,0]-c_pred[0])
    obs_s, pred_s = obs[np.argsort(ang_obs)], pred[np.argsort(ang_pred)]
    diff = np.abs(obs_s - pred_s)
    return diff.sum()


obs = load_data(CSV_FILE)
tvals = np.linspace(T_MIN, T_MAX, len(obs))

print("Running global optimization (differential evolution)...")
res_de = differential_evolution(lambda p: align_and_l1(p, obs, tvals),
                                BOUNDS, maxiter=40, popsize=12, polish=True)
print("DE result:", res_de.x, "L1:", res_de.fun)

print("Refining locally with Nelder-Mead...")
res_local = minimize(lambda p: align_and_l1(p, obs, tvals),
                     x0=res_de.x, method='Nelder-Mead',
                     options={'maxiter':1500,'xatol':1e-8,'fatol':1e-8})

theta_deg, M_est, X_est = res_local.x
print("\nâœ… Final parameters:")
print(f"Theta (deg) = {theta_deg:.6f}")
print(f"M = {M_est:.8f}")
print(f"X = {X_est:.6f}")
print(f"Objective (L1 sum) = {res_local.fun:.3f}")

# ===========================================
# STEP 4: Plot observed vs predicted curve
# ===========================================
pred = predict(np.deg2rad(theta_deg), M_est, X_est, tvals)
c_obs, c_pred = obs.mean(axis=0), pred.mean(axis=0)
ang_obs = np.arctan2(obs[:,1]-c_obs[1], obs[:,0]-c_obs[0])
ang_pred = np.arctan2(pred[:,1]-c_pred[1], pred[:,0]-c_pred[0])
obs_s, pred_s = obs[np.argsort(ang_obs)], pred[np.argsort(ang_pred)]

plt.figure(figsize=(8,6))
plt.plot(obs_s[:,0], obs_s[:,1], '.', label='Observed', alpha=0.7)
plt.plot(pred_s[:,0], pred_s[:,1], '.', label='Predicted', alpha=0.7)
plt.legend(); plt.xlabel('x'); plt.ylabel('y')
plt.title('Observed vs Predicted Curve')
plt.grid(True)
plt.show()

# ===========================================
# STEP 5: Generate LaTeX + Desmos strings
# ===========================================
latex = r"""\begin{aligned}
x(t)&=t\cos(%.6f^\circ) - e^{%.8f|t|}\sin(0.3t)\sin(%.6f^\circ) + %.6f\\[4pt]
y(t)&=42 + t\sin(%.6f^\circ) + e^{%.8f|t|}\sin(0.3t)\cos(%.6f^\circ)
\end{aligned}""" % (theta_deg, M_est, theta_deg, X_est, theta_deg, M_est, theta_deg)

desmos = "( t*cos(%.6f) - e^(%.8f*abs(t))*sin(0.3*t)*sin(%.6f) + %.6f, 42 + t*sin(%.6f) + e^(%.8f*abs(t))*sin(0.3*t)*cos(%.6f) )" % (
    np.deg2rad(theta_deg), M_est, np.deg2rad(theta_deg), X_est, np.deg2rad(theta_deg), M_est, np.deg2rad(theta_deg)
)

print("\nðŸ“„ LaTeX equation (copy to README):\n")
print(latex)
print("\nðŸ“„ Desmos equation (copy to Desmos parametric input):\n")
print(desmos)

# Save to text files for download
with open("submission_latex.txt", "w") as f: f.write(latex)
with open("submission_desmos.txt", "w") as f: f.write(desmos)

print("\nâœ… Saved:")
print(" - submission_latex.txt")
print(" - submission_desmos.txt")
print(" - You can download them from the left file pane.")
```
